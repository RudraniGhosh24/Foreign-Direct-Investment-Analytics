```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("FDI_data.csv", index_col=0)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Yearwise analysis
df_yearwise = df.transpose()
df_yearwise.plot(kind='line', figsize=(15, 8))
plt.title('Year-wise Investment Analysis from 2000-01 to 2016-17')
plt.xlabel('Financial Year')
plt.ylabel('FDI')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df['Total'] = df.sum(axis=1)

df_sorted = df.sort_values(by='Total', ascending=False)

# Display the top 10 sectors contributing the most to FDI
top_sectors = df_sorted.head(10)  
print("Top 10 Contributing Sectors to FDI in between 2000-01 and 2016-17:\n")
print(top_sectors[['Total']])

plt.figure(figsize=(12, 6))
top_sectors['Total'].plot(kind='bar', color='skyblue')
plt.title('Top 10 Contributing Sectors to FDI in between 2000-01 and 2016-17')
plt.xlabel('Sector')
plt.ylabel('Total FDI')
plt.xticks(rotation=90)
plt.show()
```

```python


# KPI 1: Total FDI Inflow
total_fdi = df.sum(axis=0)

# KPI 2: Sector-wise FDI Share
fdi_share_percentage = df.divide(df.sum(axis=1), axis=0) * 100

# KPI 3: Yearly Growth Rate
yearly_growth_rate = df.pct_change().mean() * 100

# KPI 4: Top Contributing Sectors
df['Total'] = df.sum(axis=1)
top_sectors = df.nlargest(5, 'Total')  # Adjust the number based on your preference

# KPI 5: Sector-wise Average Annual Growth
avg_annual_growth = df.pct_change().mean(axis=1) * 100

# KPI 6: Concentration Index
concentration_index = df.sum(axis=1).std()

# KPI 7: Volatility in FDI
fdi_volatility = df.std(axis=1)

# Display KPIs
print("\nKPI 1: Total FDI Inflow")
print(total_fdi)

print("\nKPI 2: Sector-wise FDI Share (%)")
print(fdi_share_percentage)

print("\nKPI 3: Yearly Growth Rate (%)")
print(yearly_growth_rate)

print("\nKPI 4: Top Contributing Sectors")
print(top_sectors[['Total']])

print("\nKPI 5: Sector-wise Average Annual Growth (%)")
print(avg_annual_growth)

print("\nKPI 6: Concentration Index")
print(concentration_index)

print("\nKPI 7: Volatility in FDI")
print(fdi_volatility)


```

```python
# Plotting Cumulative FDI over Time 
cumulative_fdi = df.drop(columns=['Total']).cumsum()
plt.figure(figsize=(15, 8))
for sector in cumulative_fdi.index:
    plt.plot(cumulative_fdi.columns, cumulative_fdi.loc[sector], label=sector)

plt.title('Cumulative FDI Over Time (2000-01 to 2016-17)')
plt.xlabel('Financial Year')
plt.ylabel('Cumulative FDI')
plt.xticks(rotation=90)
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
